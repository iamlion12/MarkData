```python
from IPython.display import display, clear_output
import ipywidgets as widgets
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
import os

%matplotlib inline
```

## Инструкция:

* В папку images надо залить изображения
* Прожимаете следующие 2 ячейки
* Во второй ячейке будут показываться изображения и 3 кнопки
* Кнопки Да\Нет для отметок, Save Data - сохраняет метки для файлов

```python
marks = {}
files = os.listdir(os.path.join('images/'))
name = ''
```

```python
button1 = widgets.Button(description="Yes")
button2 = widgets.Button(description="No")
button3 = widgets.Button(description="Save data")

display(button1, button2, button3)

def on_button_clicked(b):
    marks[name] = 1
    files.remove(name)
    dsp()

def on_button_clicked2(b):
    marks[name] = 2
    files.remove(name)
    dsp()

def save():
    if not os.path.isfile('data.txt'):
        with open('data.txt', 'w') as File:
            File.write(str(marks))
    else:
        with open('data.txt', 'r') as File:
            marks.update(dict(File.read()))
            
        with open('data.txt', 'w') as File:
            File.write(str(marks))

def on_button_clicked3(b):
    save()
    
def dsp():
    global name
    
    if files == []:
        print("All work is done. Chill out.")
        save()
        clear_output()
        
    else:
        img_name = files[0]
        name = img_name
        img = imread(os.path.join('images/'+img_name))
        plt.imshow(img)

button1.on_click(on_button_clicked)
button2.on_click(on_button_clicked2)
dsp()
```
